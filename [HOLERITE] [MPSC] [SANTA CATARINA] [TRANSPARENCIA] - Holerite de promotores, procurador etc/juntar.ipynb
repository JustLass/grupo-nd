```python
import os
import pandas as pd
```

```python
caminho = os.path.join('data', f'2020-1.xlsx')
df_1 = pd.read_excel(caminho, skiprows=3)
df_1.columns
```

```python
colunas = [
    'Cargo(s) / Função(ões)', 'Lotação(ões)',
    'Remuneração do Cargo Efetivo',
    'Outras Verbas Remuneratórias, Legais ou Judiciais',
    'Função de Confiança ou Cargo em Comissão',
    'Férias (1/3 constitucional)', 'Abono de Permanência',
    'Verbas Indenizatórias', 'Outras Remunerações Temporárias',
]

colunas2 = [
    'Cargo(s) / Função(ões)', 'Lotação(ões)',
    'Remuneração do Cargo Efetivo',
    'Função de Confiança ou Cargo em Comissão',
    'Férias (1/3 constitucional)', 'Abono de Permanência',
    'Verbas Indenizatórias', 'Outras Remunerações Temporárias',
]

dfs = []
for x in range(2020,2026):
    for i in range(1, 12):
        
        try:
            data = f'{x}-{i}'
            caminho = os.path.join('data', f'{x}-{i}.xlsx')
            try:
                df = pd.read_excel(caminho, skiprows=3)[colunas]
            except:
                print('coluna 2')
                df = pd.read_excel(caminho, skiprows=3)[colunas2]
            
            df['Referencia'] = data
            dfs.append(df)
            print(f'{data} adicionado')
        except Exception as e:
            print(f'{data} nao adicionado', type(e))
```

```python
df_final = pd.concat(dfs, ignore_index=True)
```
