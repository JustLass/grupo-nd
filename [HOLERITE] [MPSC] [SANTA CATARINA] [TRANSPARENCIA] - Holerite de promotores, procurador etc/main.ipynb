```python
import time
from pathlib import Path
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# O 'download_dir' definido na Seção 1
download_dir = Path("C:\\Users\\rafae_9y1k772\\Documents\\Grupo ND\\Webscraping\\[HOLERITE] [MPSC] [SANTA CATARINA] [TRANSPARENCIA] - Holerite de promotores, procurador etc\\data" ) 

def renomear_arquivo_apos_download(driver, elemento_download, nome_final_arquivo, timeout=30):
    """
    Clica no elemento, espera o download e renomeia o arquivo.
    """
    # Garante que o diretório exista
    download_dir.mkdir(parents=True, exist_ok=True)
    
    # 1. Obter os nomes dos arquivos ANTES do download
    arquivos_antes = set(p.name for p in download_dir.glob('*'))

    # 2. Clicar no botão de download
    WebDriverWait(driver, timeout).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, elemento_download))
    ).click()

    # 3. Esperar até que um novo arquivo apareça (ou o download temporário)
    tempo_inicio = time.time()
    novo_arquivo_path = None
    while time.time() - tempo_inicio < timeout:
        arquivos_depois = set(p.name for p in download_dir.glob('*'))
        novos_arquivos = arquivos_depois - arquivos_antes
        
        # Encontra o arquivo que não estava lá antes
        if novos_arquivos:
            # Pega o primeiro novo arquivo (pode precisar de lógica mais complexa se houver vários)
            novo_arquivo_nome = next(iter(novos_arquivos)) 
            novo_arquivo_path = download_dir / novo_arquivo_nome
            
            # 4. Esperar que o download seja concluído (sem extensão temporária)
            if not novo_arquivo_path.suffix in ['.crdownload', '.part']:
                break
            
        time.sleep(1)
    
    # Verifica se o arquivo foi encontrado e baixado
    if novo_arquivo_path and novo_arquivo_path.is_file():
        # 5. Renomear o arquivo
        nome_final = download_dir / nome_final_arquivo
        novo_arquivo_path.rename(nome_final)
        print(f"Arquivo renomeado de '{novo_arquivo_path.name}' para '{nome_final_arquivo}'")
        return True
    else:
        print("Erro: Não foi possível encontrar o arquivo baixado.")
        return False

# --- Exemplo de Uso ---
# Renomeia o primeiro download
```
