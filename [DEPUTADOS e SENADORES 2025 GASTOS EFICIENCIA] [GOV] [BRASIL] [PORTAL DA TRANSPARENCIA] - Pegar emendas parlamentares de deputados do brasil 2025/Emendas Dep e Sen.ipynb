```python
# Exemplo de iteração digitando o nome completo
# IMPORTANTE: Substitua 'SELETOR_DO_INPUT' pelo seletor correto do campo de busca na página
input_selector = "token-input-autorEmenda" # Ex: "input-busca" ou By.ID, "id-do-campo"
dfs_final = []
for nome in deputados_em_exercicio:
    try:
        # Localiza o campo de busca (ajuste o método de busca conforme necessário: By.ID, By.NAME, etc.)
        # Aqui estou assumindo um ID fictício para o exemplo
        
        driver.find_element(By.CSS_SELECTOR, "#btn-nome-do-autor-da-emenda-3").click()

        campo_busca = driver.find_element(By.ID, input_selector)
        
        campo_busca.clear()
        
        # Digita o nome completo
        campo_busca.send_keys(nome)
        selecionar = driver.find_element(By.CSS_SELECTOR, "li.token-input-dropdown-item2").click() 
        adicionar = driver.find_element(By.CSS_SELECTOR, "#id-box-filtro > div > div > ul > li:nth-child(2) > div > div:nth-child(4) > div.gaveta.gaveta-aberta > div > div.gaveta__corpo > button").click()
        consultar = driver.find_element(By.CSS_SELECTOR, "#id9 > div.d-flex.justify-content-end.pb-2 > div:nth-child(1) > button").click()
        time.sleep(2)
        dfs = []
        while True:
            try:
                print('ler html')
                sio = StringIO(driver.page_source)
                df = pd.read_html(sio)[0]
                print('Adicionando lista')
                dfs.append(df)
                print('proximo')
                driver.find_element(By.CSS_SELECTOR, "#lista_next > button").click()
                time.sleep(2)
            except Exception as e:
                print(f"Acabou paginação")
                break
        print(len(dfs), 'dfs')
        df_base = pd.concat(dfs)
        df_base['Autor da emenda'] = nome
        dfs_final.append(df_base)
        driver.get('https://portaldatransparencia.gov.br/emendas/consulta?paginacaoSimples=true&tamanhoPagina=&offset=&direcaoOrdenacao=asc&de=2025&ate=2025&colunasSelecionadas=linkDetalhamento%2Cano%2CtipoEmenda%2Cautor%2CnumeroEmenda%2CpossuiApoiadorSolicitante%2ClocalidadeDoGasto%2Cfuncao%2Csubfuncao%2Cprograma%2Cacao%2CplanoOrcamentario%2CcodigoEmenda%2CvalorEmpenhado%2CvalorLiquidado%2CvalorPago%2CvalorRestoInscrito%2CvalorRestoCancelado%2CvalorRestoPago')
        
        time.sleep(2) # Aguarda processamento ou visualização
        
    except Exception as e:
        print(f"Erro ao processar {nome}: {e}")
```
