```python
import os
import pandas as pd
import glob

base_path = 'microdados downloads'
search_pattern = os.path.join(base_path, '**', 'MICRODADOS_CADASTRO_CURSOS_*.CSV')
files = glob.glob(search_pattern, recursive=True)

dfs = []

print(f"Encontrados {len(files)} arquivos.")

for file_path in files:
    print(f"Lendo: {file_path}")
    try:
        # Adicionado dtype=str para ler tudo como texto
        df = pd.read_csv(file_path, encoding='latin-1', sep=';', dtype=str, low_memory=False)
        dfs.append(df)
    except Exception as e:
        print(f"Erro ao ler {file_path}: {e}")

if dfs:
    df_cursos_completo = pd.concat(dfs, ignore_index=True)
    print("Leitura concluída!")
    display(df_cursos_completo.head())
else:
    print("Nenhum arquivo encontrado.")
```

```python
columns = ['NU_ANO_CENSO', 'NO_REGIAO', 'SG_UF', 'CO_UF', 'NO_MUNICIPIO', 'CO_MUNICIPIO', 'TP_MODALIDADE_ENSINO', 'QT_ING', 'QT_MAT','QT_ING_FEM', 'QT_ING_MASC','QT_ING_BRANCA','QT_ING_PRETA', 'QT_ING_PARDA','QT_ING_AMARELA', 'QT_ING_INDIGENA', 'QT_ING_CORND',
                    'QT_ING_0_17','QT_ING_18_24','QT_ING_25_29','QT_ING_30_34','QT_ING_35_39','QT_ING_40_49','QT_ING_50_59','QT_ING_60_MAIS' ]
df = df_cursos_completo[columns].copy()
```

```python
def export(df,name):
    df.to_csv(f'./correcao/{name}.csv', index=False)
```

```python
# Dados de 2024
df_2024 = df_sc[df_sc['NU_ANO_CENSO'] == '2024'].copy()

df_2014 = df_sc[df_sc['NU_ANO_CENSO'] == '2014'].copy()
```

```python
# 1

# Filtrar ano 2024
df_2024 = df_sc[df_sc['NU_ANO_CENSO'] == '2024'].copy()
df_2024['QT_MAT'] = pd.to_numeric(df_2024['QT_MAT'], errors='coerce')

# Agrupar por município e modalidade
municipio = (
    df_2024.groupby(['NO_MUNICIPIO', 'MODALIDADE_NOME'])['QT_MAT'].sum()
    .reset_index()
    .pivot(index='NO_MUNICIPIO', columns='MODALIDADE_NOME', values='QT_MAT')
    .rename_axis(columns=None)
    .fillna(0)
)

# Calcular total e percentuais
municipio['Total'] = municipio['EaD'] + municipio['Presencial']
municipio['% EaD'] = (municipio['EaD'] / municipio['Total'] * 100).round(2)
municipio['% Presencial'] = (municipio['Presencial'] / municipio['Total'] * 100).round(2)

municipio = municipio.sort_values('NO_MUNICIPIO', ascending=True).reset_index()
display(municipio)
```

```python
# Mancha expansao

# Converter para numérico
df_sc['NU_ANO_CENSO'] = pd.to_numeric(df_sc['NU_ANO_CENSO'], errors='coerce')

# Primeiro ano com ensino superior por município
primeiro_ano = (
    df_sc.groupby('NO_MUNICIPIO')['NU_ANO_CENSO']
    .min()
    .reset_index()
    .rename(columns={'NU_ANO_CENSO': 'ANO_INICIO_ENSINO_SUPERIOR'})
    .sort_values('ANO_INICIO_ENSINO_SUPERIOR')
)

display(primeiro_ano)

# Quantidade de municípios por ano de início
qtd_por_ano = (
    primeiro_ano.groupby('ANO_INICIO_ENSINO_SUPERIOR')
    .size()
    .reset_index(name='QTD_MUNICIPIOS')
)

display(qtd_por_ano)

export(primeiro_ano, 'topico_2 Mancha da expansao')
export(qtd_por_ano, 'topico_2 detalhes da expansao')
```

```python

# 3 - Percentual EaD vs Presencial por UF (Brasil, 2024)
df_br_2024 = df[df['NU_ANO_CENSO'] == '2024'].copy()
df_br_2024['QT_MAT'] = pd.to_numeric(df_br_2024['QT_MAT'], errors='coerce')
df_br_2024['MODALIDADE_NOME'] = df_br_2024['TP_MODALIDADE_ENSINO'].map({'1': 'Presencial', '2': 'EaD'})

uf_modalidade = (
    df_br_2024.groupby(['SG_UF', 'MODALIDADE_NOME'])['QT_MAT'].sum()
    .reset_index()
    .pivot(index='SG_UF', columns='MODALIDADE_NOME', values='QT_MAT')
    .rename_axis(columns=None)
    .fillna(0)
)

uf_modalidade['Total'] = uf_modalidade['EaD'] + uf_modalidade['Presencial']
uf_modalidade['% EaD'] = (uf_modalidade['EaD'] / uf_modalidade['Total'] * 100).round(2)
uf_modalidade['% Presencial'] = (uf_modalidade['Presencial'] / uf_modalidade['Total'] * 100).round(2)

uf_modalidade = uf_modalidade.sort_values('% EaD', ascending=False).reset_index()
display(uf_modalidade)
```

```python
# Perfil etário dos ingressantes por modalidade - SC 2024
df_idade = df_sc[df_sc['NU_ANO_CENSO'] == 2024].copy()

ing_cols = ['QT_ING_0_17', 'QT_ING_18_24', 'QT_ING_25_29', 'QT_ING_30_34',
            'QT_ING_35_39', 'QT_ING_40_49', 'QT_ING_50_59', 'QT_ING_60_MAIS']
df_idade[ing_cols] = df_idade[ing_cols].apply(pd.to_numeric, errors='coerce')

# Agrupar por modalidade
grupo = df_idade.groupby('MODALIDADE_NOME')[ing_cols].sum()

# Criar faixas
grupo['Jovens (até 24)'] = grupo['QT_ING_0_17'] + grupo['QT_ING_18_24']
grupo['Adultos (25 a 49)'] = grupo['QT_ING_25_29'] + grupo['QT_ING_30_34'] + grupo['QT_ING_35_39'] + grupo['QT_ING_40_49']
grupo['Senior (50+)'] = grupo['QT_ING_50_59'] + grupo['QT_ING_60_MAIS']

faixas = ['Jovens (até 24)', 'Adultos (25 a 49)', 'Senior (50+)']
total = grupo[faixas].sum(axis=1)

perfil = grupo[faixas].div(total, axis=0) * 100
perfil = perfil.round(2)
perfil.columns = ['% Jovens (até 24)', '% Adultos (25 a 49)', '% Senior (50+)']
perfil.index.name = 'MODALIDADE'

display(perfil)
```

```python
# 6

# 6 - Dependência EaD por porte do município (SC, 2024)
pop = pd.read_excel('SC_POP2022.xlsx')
pop = pop.rename(columns={'NOME DO MUNICÍPIO': 'NO_MUNICIPIO', 'POPULAÇÃO': 'POPULACAO'})

# Cruzar dados INEP 2024 com população
df_2024_sc = df_sc[df_sc['NU_ANO_CENSO'] == 2024].copy()
df_porte = df_2024_sc.merge(pop[['NO_MUNICIPIO', 'POPULACAO']], on='NO_MUNICIPIO', how='left')

# Classificar porte
def classificar_porte(pop):
    if pop < 20000:
        return '1. Pequena (até 20 mil habitantes)'
    elif pop < 50000:
        return '2. Média-Pequena (20 a 50 mil)'
    elif pop < 100000:
        return '3. Média-Grande (50 a 100 mil)'
    else:
        return '4. Grande Centro (mais de 100 mil)'

df_porte['PORTE'] = df_porte['POPULACAO'].apply(classificar_porte)

# Agrupar por porte e modalidade
porte_mod = (
    df_porte.groupby(['PORTE', 'MODALIDADE_NOME'])['QT_MAT'].sum()
    .reset_index()
    .pivot(index='PORTE', columns='MODALIDADE_NOME', values='QT_MAT')
    .rename_axis(columns=None)
    .fillna(0)
)

porte_mod['Total'] = porte_mod['EaD'] + porte_mod['Presencial']
porte_mod['% EaD'] = (porte_mod['EaD'] / porte_mod['Total'] * 100).round(2)
porte_mod['% Presencial'] = (porte_mod['Presencial'] / porte_mod['Total'] * 100).round(2)

display(porte_mod[['% EaD', '% Presencial']])
```

```python
# 6 - Dependência EaD por porte do município (SC, 2024)
pop = pd.read_excel('SC_POP2022.xlsx')
pop = pop.rename(columns={'NOME DO MUNICÍPIO': 'NO_MUNICIPIO', 'POPULAÇÃO': 'POPULACAO'})
pop['POPULACAO'] = (pop['POPULACAO'] * 1000).astype(int)

# Cruzar dados INEP 2024 com população
df_2024_sc = df_sc[df_sc['NU_ANO_CENSO'] == 2024].copy()
df_porte = df_2024_sc.merge(pop[['NO_MUNICIPIO', 'POPULACAO']], on='NO_MUNICIPIO', how='left')

# --- Exploração ---
print(f"Total de registros (cursos) em 2024 SC: {len(df_2024_sc)}")
print(f"Municípios no INEP: {df_2024_sc['NO_MUNICIPIO'].nunique()}")
print(f"Municípios no Censo Pop: {pop['NO_MUNICIPIO'].nunique()}")
print(f"Registros sem match de população: {df_porte['POPULACAO'].isna().sum()}")

# Municípios que não casaram
sem_match = df_porte[df_porte['POPULACAO'].isna()]['NO_MUNICIPIO'].unique()
if len(sem_match) > 0:
    print(f"\nMunicípios sem correspondência ({len(sem_match)}):")
    for m in sem_match:
        print(f"  - {m}")

# Distribuição por porte
def classificar_porte(p):
    if p < 20000:
        return '1. Pequena (até 20 mil habitantes)'
    elif p < 50000:
        return '2. Média-Pequena (20 a 50 mil)'
    elif p < 100000:
        return '3. Média-Grande (50 a 100 mil)'
    else:
        return '4. Grande Centro (mais de 100 mil)'

df_porte['PORTE'] = df_porte['POPULACAO'].apply(classificar_porte)

# Quantos municípios em cada faixa
mun_por_porte = df_porte.groupby('PORTE')['NO_MUNICIPIO'].nunique().reset_index(name='QTD_MUNICIPIOS')
print("\nMunicípios por faixa de porte:")
display(mun_por_porte)

# Total de matrículas por porte
mat_por_porte = df_porte.groupby('PORTE')['QT_MAT'].sum().reset_index()
print("Matrículas por porte:")
display(mat_por_porte)

# Resultado final: % EaD vs Presencial por porte
porte_mod = (
    df_porte.groupby(['PORTE', 'MODALIDADE_NOME'])['QT_MAT'].sum()
    .reset_index()
    .pivot(index='PORTE', columns='MODALIDADE_NOME', values='QT_MAT')
    .rename_axis(columns=None)
    .fillna(0)
)

porte_mod['Total'] = porte_mod['EaD'] + porte_mod['Presencial']
porte_mod['% EaD'] = (porte_mod['EaD'] / porte_mod['Total'] * 100).round(2)
porte_mod['% Presencial'] = (porte_mod['Presencial'] / porte_mod['Total'] * 100).round(2)

print("\nResultado final:")
display(porte_mod)
```

```python
# 7 - Faculdades privadas dominam o EaD (SC, 2024)
df7 = df_cursos_completo[df_cursos_completo['SG_UF'] == 'SC'].copy()
df7 = df7[df7['NU_ANO_CENSO'] == '2024'].copy()
df7['QT_MAT'] = pd.to_numeric(df7['QT_MAT'], errors='coerce')
df7['MODALIDADE'] = df7['TP_MODALIDADE_ENSINO'].map({'1': 'Presencial', '2': 'EaD'})
df7['TIPO'] = df7['TP_REDE'].map({'1': 'PÚBLICA', '2': 'PRIVADA'})

# Presencial
presencial = df7[df7['MODALIDADE'] == 'Presencial'].groupby('TIPO')['QT_MAT'].sum().reset_index()
presencial.columns = ['TIPO', 'PRESENCIAL']

print("PRESENCIAL (QT_MAT em 2024 em SC)")
display(presencial.sort_values('PRESENCIAL', ascending=True))

# EaD
ead = df7[df7['MODALIDADE'] == 'EaD'].groupby('TIPO')['QT_MAT'].sum().reset_index()
ead.columns = ['TIPO', 'EAD']
print("\nEaD (QT_MAT em 2024 em SC)")
display(ead.sort_values('EAD', ascending=True))
```
